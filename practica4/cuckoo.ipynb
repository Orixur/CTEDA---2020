```python
import math
import copy
import pandas as pd
```

```python
class UnhasheableError(Exception):
    pass

class Coco():
    def __init__(self, size:int):
        self._data = [None] * size
        self._data_aux = [None] * size
        self._history_data = []
        self._history_data_aux = []
        self._iter_count_limit = 1000000
    
    @property
    def iter_count_limit(self):
        return self._iter_count_limit
    
    @iter_count_limit.setter
    def iter_count_limit(self, value):
        self._iter_count_limit = value
    
    @property
    def data(self):
        return self._data

    @data.setter
    def data(self, value):
        self._data = value

    @property
    def data_aux(self):
        return self._data_aux

    @data_aux.setter
    def data_aux(self, value):
        self._data_aux = value
    
    @property
    def history_data(self):
        return self._history_data
    
    @history_data.setter
    def history_data(self, value):
        self._history_data = value
        
    @property
    def history_data_aux(self):
        return self._history_data_aux
    
    @history_data_aux.setter
    def history_data_aux(self, value):
        self._history_data_aux = value
        
    def show_history(self):
        df_data =  pd.DataFrame()
        for n, h in enumerate(self._history_data):
            df_data[f'iter-{n}'] = h
        
        df_aux =  pd.DataFrame()
        for n, h in enumerate(self._history_data_aux):
            df_aux[f'iter-{n}'] = h
        
        return df_data, df_aux

    def add_handler(self, value):
        def add(value):
            current_hash1 = value[0] % 11
            current_hash2 = math.floor(value[0]/11) % 11
            
            aux = None
            if value[1] == 'main':
                if self._data[current_hash1] == None:
                    self._data[current_hash1] = value[0]
                else:
                    aux = (self._data[current_hash1], 'aux')
                    self._data[current_hash1] = value[0]
            elif value[1] == 'aux':
                if self._data_aux[current_hash2] == None:
                    self._data_aux[current_hash2] = value[0]
                else:
                    aux = (self._data_aux[current_hash2], 'main')
                    self._data_aux[current_hash2] = value[0]
            return aux
        
        iter_count = 0
        current = (value, 'main')
        while current is not None:
            current = add(current)
            self._history_data.append(copy.deepcopy(self._data))
            self._history_data_aux.append(copy.deepcopy(self._data_aux))
            iter_count += 1
            if iter_count >= self._iter_count_limit:
                del self._history_data
                del self._history_data_aux
                raise UnhasheableError(f'El valor {value} no se puede hashear, se interrumpirá la ejecución.')
```

## Conclusión
Este ejercicio es imposible de terminar, puesto que en el momento en el que queramos insertar por segunda vez el valor 3, se creará un búcle infinito entre los valores: 3, 5, 25 y 27.

> En este caso se seteo un limite máximo de seguridad de 1.000.000 de intentos para encontrar la ubicación adecuada en ambas listas.

El problema con este par de funciones de hash es que un valor puede llegar a tener la misma ubicación en ambos almacenamientos.

## Solución
Es posible sanear este error utilizando un segundo conjunto de hashs para recalcular los indices, usando un área de stashing (es un vector donde se almacenan los datos que no se pueden hashear, se añaden a la busqueda y eliminación), agrandando los almacenamientos, utilizando mas particiones, etc...
